# process

```python
sql08_res['pay_rate'] = sql08_res['pay_user'] / sql08_res['dau']
sql08_res['pay_rate_cate'] = sql08_res['pay_user_cate'] / sql08_res['dau_cate']
# sql08_res['pau_rate'] = sql08_res['pay_user'] / sql08_res['pay_dau']
```

```python
def data_process(data: pd, is_cate=True):
    if is_cate == True:
        data['group_col'] = data['payment_cate']
        data['pay_user_col'] = data['pay_user_cate']
        data['pay_rate'] = data['pay_user_cate'] / data['pay_dau_cate']        
    else:
        data['group_col'] = data['payment_name']
        data['pay_user_col'] = data['pay_user']
        data['pay_rate'] = data['pay_user'] / data['pay_dau']

    data_res = data.groupby(['group_col', 'month'])['pay_rate'].describe(percentiles=np.arange(0, 1.1, 0.1))
    # data_res['pay_user'] = data.groupby(['group_col', 'month'])['pay_user_col'].sum()

    return data_res
```

```python
py08_01res = data_process(sql08_res)
py08_02res = data_process(sql08_res, False)
```

```python
py08_01res.to_csv('./py08_01res.csv')
py08_02res.to_csv('./py08_02res.csv')
```
